```python
# 고객당 총구매횟수
sql = '''
select c.customer_id, count(*) '구매횟수' , email, month(rental_date) 'month'
from customer c join rental r
on c.customer_id = r.customer_id
group by customer_id
having 구매횟수 >= 30
'''
pd.read_sql(sql, db)
```

```python
# 월별 고객당 총구매횟수
sql = '''
select c.customer_id, count(*) '구매횟수' , email, month(rental_date) 'month'
from customer c join rental r
on c.customer_id = r.customer_id
group by month, c.customer_id
order by 구매횟수 desc;
'''
pd.read_sql(sql, db)
```

```python
sql = '''
select title, sum(amount) '총수입' from payment p, rental r, inventory i, film f
where p.rental_id =r.rental_id and r.inventory_id=i.inventory_id and
i.film_id=f.film_id
group by title
order by 총수입 desc;
'''
pd.read_sql(sql, db)
```

```python
sql = '''
select store_id, rating, sum(amount) 'sales' from payment p, rental r, inventory i, film f
where p.rental_id=r.rental_id and
r.inventory_id=i.inventory_Id and
i.film_Id=f.film_Id
group by rating
having store_id=1;
'''
pd.read_sql(sql, db)
```

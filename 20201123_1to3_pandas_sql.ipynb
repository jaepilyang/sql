```python
# 조용필이 구매한 제품의 이름, 주소 출력
pd.read_sql('select prodName, addr from usertbl u join buytbl b on u.userid=b.userid where name="조용필";', db)
```

```python
user_df = pd.read_sql('select * from usertbl;', db)
buy_df = pd.read_sql('select * from buytbl;', db)

merge_df = pd.merge(user_df, buy_df, on='userID')
merge_df[merge_df['prodName']=='모니터'][['name', 'addr']]
```

```python
# 전화번호가 없는 고객의 이름과 주소, 구매한 제품을 출력
pd.read_sql('select Name, addr, prodname from usertbl u left outer join buytbl b on u.userid=b.userid where mobile1 is null and prodname is not null;', db)
```

```python
# 총구매금액이 가장 큰 고객의 이름과 주소, 총구매금액
sql = '''
select Name, addr, sum(amount*price) as '총구매금액' from usertbl u join buytbl b 
on u.userid=b.userid 
group by name order by 총구매금액 desc limit 1;
'''
pd.read_sql(sql, db)
```

```python
merge_df = pd.merge(user_df, buy_df, on='userID')
merge_df['buyamount'] = merge_df['amount'] * merge_df['price']
temp_df = merge_df.groupby('userID')['buyamount'].sum().to_frame()
temp_df1 = merge_df.merge(temp_df, on='userID').sort_values('buyamount_y', ascending=False)
temp_df1[['name', 'addr', 'buyamount_y']]
```

```python
# 구매횟수가 2회 이상인 고객의 userid와 구매횟수
sql = '''
select Name, u.userid, count(amount) as '총구매횟수' from usertbl u join buytbl b 
on u.userid=b.userid 
group by name 
having 총구매횟수 >=2
order by 총구매횟수 desc;
'''
pd.read_sql(sql, db)
```

```python
# 대륙은 유럽이고 인구수는  천만 이상인 국가들을 추출
start_time  = time.time()

sql='''
select name "국명", population "인구수", continent "대륙명" from country
where continent='europe' and population >=10000000
'''
pd.read_sql(sql, db)

end_time=time.time()
excute_time = end_time-start_time
print(excute_time)
```

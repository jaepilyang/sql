```python
import pandas as pd
import MySQLdb as ms
import numpy as np
```

```python
db = ms.connect('localhost', 'root', '1234', 'sakila', charset='utf8')
```

```python
pd.read_sql('show tables;', db)
```

```python
addr_df = pd.read_sql('select * from address', db)
custm_df = pd.read_sql('select * from customer', db)
custm_df
```

```python
# 우편발송 : 고객의 이름과 주소를 출력하시오.
# 판다스
pd.merge(custm_df, addr_df, on='address_id')[['first_name', 'last_name', 'address', 'address2']]
```

```python
sql = '''
select first_name, last_name, address from customer c inner join address a
on c.address_id = a.address_id
'''
pd.read_sql(sql, db)
```

```python
# 국가명과 시이름을 매칭시켜서 출력 => 국가명 칼럼, 도시명 칼럼
sql = '''
select * from city
'''
pd.read_sql(sql, db)
```

```python
sql = '''
select country.country_id, country, city from country inner join city
on country.country_id=city.country_id
'''
pd.read_sql(sql, db)
```

```python
# sakila에서 rental 테이블에서 월별 렌탈 수를 출력
# 날짜는 오름차순으로 출력
sql = '''
select * from rental;
'''
df = pd.read_sql(sql, db)
```

```python
df['month'] = df['rental_date'].dt.month
df.head()
```

```python
df.groupby('month')['rental_date'].count()
```

```python
sql = '''
select count(rental_id) "count", left(rental_date, 7) month from rental
group by month;
'''
pd.read_sql(sql, db)
```

```python
# payment 테이블에서 월별 전체매출, 평균매출, 결제건수
sql = '''
select sum(amount) "전체매출", avg(amount) "평균매출", count(amount) "결제건수", left(payment_date, 7) month 
from payment
group by month;
'''
pd.read_sql(sql, db)
```

```python
# 판다스
df = pd.read_sql('select * from payment', db)
df['month'] = df['payment_date'].dt.month
df.head()
```

```python
df.groupby('month').agg({'amount':[np.sum, np.mean], 'payment_id':len})
```
